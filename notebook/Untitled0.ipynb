```python
import pandas as pd

df = pd.read_csv("../final_data.csv")
df.head()
```

```python
df["수급자비율"] = (df["일반수급자 인원"] / df["총인구수"]) * 100
df["범죄율"] = (df["범죄건수"] / df["총인구수"]) * 100000

df.head()
```

```python
corr = df[["수급자비율", "범죄율"]].corr().iloc[0, 1]
print("Correlation coefficient:", round(corr, 3))
```

```python
df_clean = df[~df["시도"].isin(["세종", "제주"])]

corr_clean = df_clean[["수급자비율", "범죄율"]].corr().iloc[0, 1]
print("After outlier removal:", round(corr_clean, 3))
```

```python
import statsmodels.api as sm

X = sm.add_constant(df_clean["수급자비율"])
y = df_clean["범죄율"]

model = sm.OLS(y, X).fit()
print(model.summary())
```
